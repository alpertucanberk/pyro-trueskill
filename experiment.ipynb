```python


from data.get_data import get_data
from trueskill.models.base_model import model
from pyro.optim import Adam
from pyro.infer import SVI, Trace_ELBO
import pyro


adam_params = {"lr": 0.02}
optimizer = Adam(adam_params)

outcome, sparse_game_matrix = get_data('/Users/alpercanberk/Projects/pyro-trueskill/data/ChessRatings2/primary_training_part1.csv')
guide = pyro.infer.autoguide.AutoDiagonalNormal(model)

svi = SVI(model, guide, optimizer, loss=Trace_ELBO())

losses = []
for step in range(1000):
    loss = svi.step(outcome, sparse_game_matrix)
    losses.append(loss)
    if step % 1 == 0:
        print("Elbo loss: {}".format(loss))
```
